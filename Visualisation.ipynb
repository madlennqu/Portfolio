# Activity 4

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,11)
y = x**2

fig, axes = plt.subplots(1,2)

# Now use the axes object to add stuff to plot
axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title1')

axes[1].plot(y, x, 'b')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].set_title('title2')
```

# Activity 5

```python
import matplotlib.pyplot as plt

types  = ['Action','Comedy','Drama','Horror','Other','Sci-Fi']
values = [147,456,203,87,83,63]
colors = ['red', 'blue', 'cyan', 'purple', 'yellow', 'lightgreen']
prop   = [14.15,43.89,19.54,8.37,7.99,6.06]
focus  = [0, 0, 0, 0, 0, 0.2]

# Data Visualization
plt.pie(values, colors= colors, labels=types, explode = focus)
plt.title('Movie Genres')
plt.legend(bbox_to_anchor=(1.2,1.1))
plt.tight_layout()
plt.show()
```

# Activity 6

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

# Pandas
df = pd.read_csv("countryGDP.csv")
data1 = df['GDP per Capita'].dropna() 
data2 = df['Population'].dropna()/1000 

# Data Visualization
fig, axes = plt.subplots(1,2)
axes[0].hist(data1, color='red',bins=20)
axes[0].set_xlabel('GDP')
axes[0].set_xlim([0,75000])
axes[0].set_title('GDP Level')

axes[1].hist(data2, color='blue',bins=20)
axes[1].set_xlabel('Population')
axes[1].set_xlim([0,200000])
axes[1].set_title('Population in Thousands')
```
